### 1번 문제
- L=[1, 2, 3, 4, 5]일 때 다음 각 문장을 수행한 후의 결과를 보고 납득할 만한 이유를 설명하시오.

```python
L = [1,2,3,4,5]
```

#### (1)

```python
L[1:3] = [100]
print (L)
```

##### Solution
리스트 L을 슬라이싱을 이용하여 1인덱스부터 2인덱스 까지의 값을 100으로 바꾸도록 함. 따라서 결과는 1과 2 인덱스의 값인 2,3의 자리에 100의 값이 치환되어 [1,100,4,5]가 출력됨

#### (2)

```python
L[1:3] = [(100)]
print(L)
```

#### Solution
위 (1) 문제와 같이 리스트 슬라이싱이다. 1인덱스부터 2 인덱스 까지의 값을 100으로 바꾸도록 하는 것으로, 1과 2 인덱스 값인 100, 4가 100으로 치환되어 [1,100,5]가 출력됨

#### (3)

```python
L[1:3] = 100
print(L)
```

#### Solution
리스트 슬라이싱이나 값 변경 시에는 위 문제들 처럼, [ ]괄호 안에 값을 넣어 주어야 한다. 위와 같이 단순 값만 입력하는 경우는 잘못된 문장이다.

#### (4)

```python
L[1:3] = (100)
print(L)
```

#### Solution
리스트의 값을 변경할 때에 튜플의 형태로도 가능하다. 하지만 위에는 100 하나의 원소가 들어갔으므로 튜플의 형태가 아니다. 따라서 위의 형태도 잘못된 형태로 오류가 발생한다.

#### (5)

```python
L[1:3] = (100, 101, 102)
print(L)
```

#### Solution
위는 튜프의 형태로서 리스트의 값을 변경하였기 때문에 가능하다. 가장 최근의 리스트 L은 [1,100,5]였고, 100, 5 대신에 100, 101, 102를 저장했으므로, L = [1,100,101,102] 가 된다.

#### (6)

```python
L[1:3] = [100, 101, 102]
print(L)
```

#### Solution
위의 결과로 L=[1,100,101,102]가 되었는데 여기서 1과 2의 인덱스 값인 100과 101 대신에 100, 101, 102를 대입하였으므로, L=[1,100,101,102,102]가 된다.

#### (7)

```python
L[1:3] = [(100, 101, 102)]
print(L)
```

#### Solution
위의 결과로 L=[1,100,101,102,102]가 되었는데 여기서 1과 2의 인덱스 값인 100과 101 대신에 (100,101,102)를 대입하였으므로, L=[1,(100,101,102),102,102]가 된다.

#### (8)

```python
L[1:3] = [[100, 101, 102]]
print(L)
```

#### Solution
위의 결과로 L=[1,(100,101,102),102,102]가 되었는데 여기서 1과 2의 인덱스 값인 101과 102 대신에 [100,101,102]를 대입하였으므로, L=[1,[100,101,102],102]가 된다.

### 2번 문제
- 문자열 S = 'Hello World and Python'에 대해 다음 요구사항에 대한 Python 코드를 제시하시오. 

1) 단어의 순서가 역순으로 되어진 문자열 ('Python and World Hello')을 만들고

2) 1)의 결과 문자열에 대해 메소드 split과 join을 이용하여 공백을 모두 없엔 문자열 'PythonandWorldHello'을 만드시오.

```python
S = 'Hello World and Python'
S = S.split()[::-1]
K = ''
for i in S :
    K = K + i + ' '
print (K)
```

```python
P = K.split()
P = ''.join(P)

print (P)
```

#### Solution
(1)문자열 S에 Hello World and Python을 저장하고, S에 저장된 문자열을 split()이라는 메서드를 이용하여 공백을 기준으로 단어를 추출하고, 슬라이싱을 이용하여 단어를 역순으로 문자열 S에 저장시킨다. K라는 문자열을 정의하고, K에 문자열 S의 각 인덱스에 저장된 배열과 공백을 포함시켜 저장시킨다.

(2)위의 K에 split()함수를 이용하여 단어별로 추출하고, join 함수를 이용하여 다시 문자열로 만든다.

### 4번 문제
- 다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)

s = '  first star   :   second star   :    third star  '

l = ['first star', 'second star', 'third star']

#### 1) for ~ in 문을 이용한 코드를 작성하시오.

```python
s = ' first star : second star : third star ' 
l = s.split(':')                              
j = 0                                         
for i in l:                                   
    l[j] = i.strip()                          
    j = j + 1                                 
print (l)                                 
```

#### Solution
s에 문제와 같이 문자열을 할당하고, split()함수를 이용하여 ':'을 기준으로 분리하여 l이라는 리스트에 저장한다. 다음으로 for문을 이용, 내부에서 strip()함수를 이용하여 공백을 제거한 리스트 l을 출력한다.

#### 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
l = s.split(':')
l = [i.strip() for i in l]
print (l)
```

#### Solution
위의 for~in문과 비슷한 구조이지만, 리스트 내포를 이용하여 for문을 내장시켜 코드를 간결화 한다.

### 5번 문제
- 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.


#### 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
## for~in 리터럴 방식
def addall(l) :
    sum = 0
    for i in range(len(l)) :
        sum = sum + l[i]
    return sum

print (addall([1]))
print (addall([1,2,3,4,5,6,7,8,9]))
```

```python
## 리스트 내포 방식
def addall(l):
    return sum([l[i] for i in range(len(l))])

print (addall([1]))
print (addall([1,2,3,4,5,6,7,8,9]))
```

#### 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)


```python
## for~in 리터럴 방식
def addallodd(l) :
    sum = 0
    for i in range(len(l)) :
        if l[i] % 2 :
            sum = sum + l[i] 
    return sum

print (addallodd([1]))
print (addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
## 리스트 내포 방식
def addallodd(l) : 
    return sum([l[i] for i in range(len(l)) if l[i]%2])

print (addallodd([1]))
print (addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

#### Solution
리스트 내포 방식을 이용하기 위해 값을 더해주는 내장함수인 sum을 이용하였습니다.

 - f = d.copy()와 같이 사전 d를 복사하여 f 사전에 할당하는 경우, 보통 값이 참조하고 있는 값이 변해도 객체를 공유하고 있지 않기 때문에 d=e와 같이 객체를 참조하는 것과는 다르게 f의 value 값은 변하지 않는다. 하지만 shallow copy를 수행하므로, 'high' 키에서 value 값인 리스트 [4,5,6]에서 두번째 인덱스인 5의 값을 500으로 바꾸게 되면 그 값이 같이 바뀌게 된다.

### 7번 문제
- 사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오.

  [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.

#### 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}

d1 = []                #d1이라는 빈 리스트를 생성
d1 = list(d.items())   #d1에 items()함수를 이용하여 d의 아이템들을 리스트로 뽑음
d1.sort()              #sort 함수로 정렬(알파벳 오름차순)
print (d1)             #출력
```

#### 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d1.sort(reverse = True) #디폴트로 되어 있는 reverse=False를 True로 바꾸어 줌
print (d1)              #출력
```

#### 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
def key_arrangement(k) :     #사전에서 value 값을 key로 지정하는 함수
    return k[1]
d1.sort(key=key_arrangement) #key를 value로 하여 그 기준으로 정렬
print (d1)
```

#### 4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d1.sort(key=key_arrangement, reverse = True) 
#key를 value로 하여 그 기준으로 정렬함과 동시에, reverse = True로 하여 역순으로 정렬
print (d1)
```

### [Incremental Project]
- 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.

  - 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
  - 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
  - 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
  - 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
     - http://www.naver.com
     - http://www.daum.net
     - http://www.nytimes.com

```python
# www.naver.com

import string
from urllib import request 
 
url = 'http://www.naver.com'            #www.naver.com URL에 대한 파싱
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
puretext = ""                           #빈 문자열 생성

i = 0
while (i < len(source)):                #while문을 이용해 source 문자열의 길이만큼 반복 수행
    if source[i] == '<':                #만약 '<'문자열을 만나면
        while (source[i] != '>'):       #'>'이 다시 올 때 까지 아무 것도 안한 채 while문을 수행 
            i = i + 1
    elif source[i] == '\n' :
        puretext = puretext + ' '
    else:
        puretext = puretext + source[i] #그 외의 경우, puretext에 source[i]를 저장
    i = i + 1

for i in range(len(string.punctuation)):
    puretext = puretext.replace(string.punctuation[i]," ") #문자열의 구두문자들을 공백으로 바꿈

puretext = puretext.split()         #구두문자에서 공백으로 바뀐 것들을 기준으로 puretext에 리스트로 저장

dic = dict()                        #빈 사전 생성
for a in puretext:                  
    if a not in dic:                #만약 a가 사전에 아직 없을 경우
        dic[a] = 1                  #a의 value를 1로 초기화
    else:                           #그 외의 경우(사전에 있는 단어인 경우)
        dic[a] = dic[a] + 1         #단어에 해당하는 value 값에 1을 더해줌
for i in dic:                       #각 단어의 출현 빈도를 사전 형태로 저장하여 출력
    print (str(i) + ':' + str(dic[i]))
```
