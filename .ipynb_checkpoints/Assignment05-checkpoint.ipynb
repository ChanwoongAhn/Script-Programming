### [3번 문제]
- 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    def __str__(self):
        return '[Count(step:' + str(self.step) + ')]' + str(self.count)
    
    def incr(self):
        self.count = self.count + self.step
        
    def __call__(self):
        return self.incr()
    
    def __add__(self, a):
        self.count += a
        return Counter(self.count, self.step)
        
    def __sub__(self, a):
        self.count -= a
        return Counter(self.count, self.step)
        
    def __cmp__(self, a):
        return self.count - a
    def __lt__(self, a):
        return self.count < a
    def __gt__(self, a):
        return self.count > a
    def __eq__(self, a):
        return self.count == a
    def __ne__(self, a):
        return self.count != a
print ('요구사항1 - 생성자에 count 값과 step 값을 인자로 받을 수 있다')
c = Counter(10)
d = Counter(10, 2)
print()

print ('요구사항2 - Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Count클래스 내부에 구현')
print (c)
print (d)
print ()

print ('요구사항3 - step에 주어진 증분만큼 count를 증가시키는 incr()메소드를 Count클래스 내부에 구현')
c.incr()
d.incr()
print (c)
print (d)
print ()

print ('요구사항4 - 인스턴스 객체를 직접 호출(call)할 수 있도록 메소드 구현')
print ('단, 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현')
c()
d()
print (c)
print (d)
print ()

print ('요구사항5 - (+, -) 두 개의 산술연산이 수행될 수 있도록 메소드를 구현')
c = c + 5
d = d - 5
print (c)
print (d)
print ()

print ('요구사항6 - 관계연산이 수행될 수 있도록 메소드를 구현')
print (c > 10)
print (d > 10)
print (c < 10)
print (d < 10)
print (c == 17)
print (d != 9)
```

### [4번 문제]
- 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

    - init() : 생성자에서 인자로 받은 리스트 l에 있는 리스트 구성원들을 for 문을 통해 하나씩 받아와 넣어 준다. 다음으로 append를 하여 리스트에 추가시켜준 후, MySet 클래스의 eliminate_duplicate 메소드를 호출한다.
    - eliminate_duplication() : staticmethod는 인스턴스 객체와 무관하게 클래스 이름 공간에 존재하는 메소드로서 클래스 이름을 이용하여 직접 호출할 수 있는 메소드를 말한다. 따라서 init() 생성자에서 객체 생성 없이 호출될 수 있는 것이고, 동작으로는 먼저 인자로부터 받아 온 리스트(l)로 부터 새로운 리스트 s에 for문을 이용해 넣어주는데, 단 e가 리스트 s에 없는 경우에만 넣어 준다. 그리고 다시 for문을 이용하여 인자로 받아 왔던 리스트 (l)을 초기화 해준 뒤 다시 넣는다.
    - str() : str 함수는 print문을 이용하여 출력해 줄 때 호출이 되는 함수로, for 문을 이용하여 result 안에 "원소값 + ','"을 계속 넣어주고 for 문이 끝난 후에는 마지막 ','을 제거하고 "}"을 넣어 주고 result 값을 반환 해 준다.

### [5번 문제]
- 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
  * 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
  * 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
  * 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, l):        # 요구사항 1 : 두 집합의 합집합을 반환(| 연산으로 반환하기 위해 __or__)  
        for e in l:
            if e not in self:
                self.append(e)
        return MySet(self)
    
    def __and__(self, l):       # 요구사항 2 : 두 집합의 교집합을 반환(& 연산으로 반환하기 위해 __and__)  
        s = []
        for e in self:
            if e in l:
                s.append(e)
        return MySet(s)
    
    def __sub__(self, l):       # 요구사항 3 : 두 집합의 차집합을 반환(- 연산으로 반환하기 위해 __sub__)  
        for e in self:
            if e in l:
                self.remove(e)
        return MySet(self)
    
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print (s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print (t)
    print ()
    
print ('<요구사항 1>') 
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
u = s | t
print (u)
print ()

print ('<요구사항 2>')
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
k = s & t
print (k)
print ()

print ('<요구사항 3>')
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print (u)
```

### [7번 문제]
- 이전 Assignment 3 (Assignment 4가 아님)의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하여 각 단어들에 대해 출현빈도를 사전형태({'world': 2, 'hello': 1, 'python': 1})로 저장하여 출력하는 것이었다. 이번에는 Assignment 3을 다시 확장/변형하여 다음과 같은 조건을 만족하도록 구현하시오.

* 1) 새로운 클래스 WebWordsFrequency를 정의하시오.
* 2) 생성자에 URL을 0개에서 임의의 개수를 넣을 수 있도록 생성자 인수를 가변인수로 정의하여 각각의 URL을 리스트 자료형에 유지하시오.
* 3) addUrl() 메소드를 구현하여 인스턴스를 생성한 이후에도 URL을 추가할 수 있도록 한다.
* 4) removeUrl() 메소드를 구현하여 URL을 삭제할 수 있도록 한다.
* 5) listUrls() 메소드를 구현하여 현재 등록된 모든 URL을 출력하는 기능을 추가
* 6) getWordsFrequency() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환하시오.
* 7) getMaxFreqencyWords() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전에서 가장 많이 출현한 단어 리스트를 반환하시오.
* 8) 임의의 단어 1개를 파라미터로 받는 searchUrlByWord() 메소드를 구현하여 유사도가 높은 웹 사이트를 출력하시오.

```python
# Assignment#3에서 각각의 웹사이트마다 불용어 제거등의 함수를 따로 만들지 않았기 때문에 이번 과제에서 클래스를 이용해
# 확장/변형하는 과정에서 지금까지의 코드는 참조만 하여 만들었습니다.
from urllib import request 
import string    



def FrequencyCount(url): # 1) 새로운 클래스 WebWordsFrequency를 정의
    s = request.urlopen(url)
    source_bytes = s.read()
    source = source_bytes.decode('utf-8')
    for i in range(0, source.count('<!--')):
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    length = len(source)
    s = ""
    n = 1
    for x in range(0, length):
        if n==1:
            if source[x] == '<':
                n = 0
                pass
            else:
                s = s + source[x]
        elif n==0:
            if source[x] == '>':
                n = 1
            else:
                pass
    L = s.split()

    import string
    punc = string.punctuation

    for x in range(0, len(L)):
        for y in punc:
            if L[x].find(y) != -1:
                L[x] = L[x].replace(y, "")
            
            

    count = 0
    for x in range(0, len(L)):
        if L[x] == "" or L[x] == " ":
            count = count + 1



    while(count >0):
        for x in range(0, len(L)):
            if L[x] == "":
                del L[x]
                count = count -1
                break

    word = [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 
            'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then',
            'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']



    count = 0
    for x in range(0, len(L)):
        if L[x] in word:
            count = count +1

    while(count >0):    #영어 불용어를 모두 제거하기 위해 사용
        for x in range(0, len(L)):
            if L[x] in word:
                del L[x]
                count = count -1
                break



    d = {}

    for x in range(0, len(L)):
        for y in range (0, len(L)):
            if L[x] == L[y]: 
                count = count + 1
        d[L[x]] = count
        count = 0
    
    return d
    

class WebWordsFrequency:
    l = []
    def __init__(self, *url):    #생성자
        for x in url:
            self.l.append(x)
            # 요구사항 2) <생성자에 URL을 0개에서 임의의 개수를 넣을 수 있도록 생성자 인수를 가변인수로 정의하여 
            #             각각의 URL을 리스트 자료형에 유지하시오.> 만족하기 위해 for in 문 을 사용
            
    # 요구사항 3) addUrl() 메소드를 구현하여 인스턴스를 생성한 이후에도 URL을 추가할 수 있도록 한다.
    def addUrl(self, url):
        self.l.append(url)
        
    # 요구사항 4) removeUrl() 메소드를 구현하여 URL을 삭제할 수 있도록 한다.
    def removeUrl(self, url):
        self.l.remove(url)
        
    # 요구사항 5) listUrls() 메소드를 구현하여 현재 등록된 모든 URL을 출력하는 기능을 추가
    def listUrls(self):     
        for x in self.l:
            print (x)
            
    # 요구사항 6)  getWordsFrequency() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환
    def getWordsFrequency(self):
        total_d = {}
        for x in self.l:
            d = FrequencyCount(x)
            for x in d:
                if x in total_d:
                    total_d[x] = total_d[x] + d[x]
                else:
                    total_d[x] = d[x]
        
        
        if total_d == {}:
            return None
        else:
            return total_d
    # 요구사항 7) getMaxFreqencyWords() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 
    #            단어 출현 빈도 사전에서 가장 많이 출현한 단어 리스트를 반환
    def getMaxFrequencyWords(self):
        l2 = []
        d = self.getWordsFrequency()

        
        max_word = ""
        max_count = 0
        
        for x in d:
            if d[x] > max_count:
                max_count = d[x]
                max_word = x
        for x in d:
            if max_count == d[x]:
                l2.append(x)
        if l2 == []:
            return None
        else:
            return l2
```

### [8, 9번 문제]
- 위 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse = True):
        total_d = {}
        for x in self.l:
            d = FrequencyCount(x)
            for x in d:
                if x in total_d:
                    total_d[x] = total_d[x] + d[x]
                else:
                    total_d[x] = d[x]
        
        def key_priority(a) :
            return a[1]
        k = list(total_d.items())
        k.sort(key = key_priority, reverse = True);
        return k
        
w1.listUrls()
w1.removeUrl('http://www.cnn.com')
w1.removeUrl('http://www.times.com')
w1.removeUrl('https://github.com')
w1.removeUrl('http://stackoverflow.com')
```
